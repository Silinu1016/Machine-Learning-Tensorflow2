```python
import tensorflow as tf
import numpy as np
import pandas as pd
tf.random.set_seed(777)

xy=pd.read_csv('framingham.csv')
xy=xy.values
xy[np.isnan(xy)]=0
test_line=int(len(xy)*0.7)
np.random.shuffle(xy)
x_data=xy[:test_line,0:-1]
y_data=xy[:test_line,[-1]] 
x_test=xy[test_line:,0:-1]
y_test=xy[test_line:,[-1]] 
tf.model = tf.keras.Sequential() 
tf.model.add(tf.keras.layers.Dense(units=1, input_dim=15)) 
 # use sigmoid activation for 0~1 problem 
tf.model.add(tf.keras.layers.Activation('relu')) 
 
tf.model.compile(loss='binary_crossentropy', optimizer=tf.keras.optimizers.SGD(lr=0.1), metrics=['accuracy']) 
tf.model.summary() 

 
history = tf.model.fit(x_data, y_data, epochs=100) 
y_predict = tf.model.predict(x_data) 
print("Predict: {0}".format(y_predict)) 

 
# evaluating model 
evaluate = tf.model.evaluate(x_test, y_test) 
print("loss: {0}, accuracy: {1}".format(evaluate[0], evaluate[1])) 

```
